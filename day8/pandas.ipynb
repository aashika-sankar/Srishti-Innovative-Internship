```python
import pandas as pd
```

```python
#from dictionary
data={'Name':['Tom','nick','krish','jack'],'Age':[20,21,19,18]}
df=pd.DataFrame(data)
print('dataframe:',df)
```

```python
#from list of lists
lst=[['tom',10],['nick',15],['juli',14]]
df=pd.DataFrame(lst,columns=['Name','Age'])
print(df)
```

```python
#inspecting data
print('firstrows 5',df.head())
print('last 5 rows',df.tail())
print('info',df.info)
print('describe',df.describe())
```

```python
#selecting data
print('Name column',df['Name'])
print('Name and Age',df[['Name','Age']])
```

```python
#row selection
print('first row',df.loc[0])
print('first row',df.iloc[0])
print('first 2 rows',df.iloc[0:2])
```

```python
#filtering data
print('rows where age',df[df['Age']>13])
```

```python
#adding and removig
df['Genter']=['M','M','F'] 
print('after add new coloum',df)
df.drop('Genter',axis=1,inplace=True) 
print('after remove',df)
```

```python
data={'Name':['Tom','nick','krish','jack'],'Age':[20,21,19,None]}
df=pd.DataFrame(data)
print(df)
df.fillna(0,inplace=True)
print(df)
```

```python
data={'Name':['Tom','nick','krish','jack'],'Age':[20,21,19,None]}
df=pd.DataFrame(data)
print(df)
print(df.isnull().sum())
```

```python
#handling missing values
data={'Name':['Tom','nick','krish','jack'],'Age':[20,21,19,None]}
df=pd.DataFrame(data)
print(df)
df.dropna(inplace=True)
print(df)
```

```python
#grouping and aggragation
grouped=df.groupby('Name')['Age'].mean()
print(grouped)
```

```python
#grouping and merging
data={'Name':['Tom','nick','krish','jack'],'Age':[20,21,19,None]}
df1=pd.DataFrame(data)
data2={'Name':['Liya','Noah','Diya','Roopa'],'Age':[20,21,19,None]}
df2=pd.DataFrame(data2)
merged=pd.merge(df,df2,on='Age')
print(merged)
```

```python
#read csv
df.to_csv('data.csv',index=False)
df=pd.read_csv('data.csv')
print(df)
```

```python
#useful functions
print('sort the data',df.sort_values('Age'))
print('change column',df['Age'].apply(lambda x:x+2))
print('value counts',df['Name'].value_counts)
```
