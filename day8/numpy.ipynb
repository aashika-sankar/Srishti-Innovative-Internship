```python
import numpy as np
```

```python
arr=np.array([1,2,3])
arr_2d=np.array([[1,2,3],[4,5,6]])
print('array\n',arr)
print('2d array\n',arr_2d)
```

```python
# Special Arrays
print('zero array',np.zeros((3, 3)))
print('ones array',np.ones((2, 4)))
print('Identity array',np.eye(3))
print('arange array',np.arange(0, 10, 2))
print('linspace array',np.linspace(0, 1, 5))
print('random array',np.random.rand(3, 2))
print('random array',np.random.randn(3, 2))
```

```python
#array attributes
arr=np.array([[1,2,3],[4,5,6]])
print(arr.shape)
print(arr.dtype)
print(arr.ndim)
print(arr.size)
```

```python
#array operations
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
print('addition',arr1+arr2)
print('subtraction',arr1-arr2)
print('multiplication',arr1*arr2)
print('division',arr1/arr2)
print('sin',np.sin(arr1))
```

```python
#matrix multiplicatioon
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print('matrix multiplication',a.dot(b))
print('matrix multiplication using @',a@b)
```

```python
#aggregation
arr=np.array([[1,2,3],[4,5,6]])
print('sum',arr.sum())
print('min',arr.min())
print('max',arr.max())
print('mean',arr.mean())
print('std',arr.std())
```

```python
#indexing and slicing
arr=np.array([1,2,3,4,5,6])
print('first element',arr[0])
print('last element',arr[-1])
print('first 3 elements',arr[:3])
print('last 3 elements',arr[-3:]) 
print('every other element',arr[::2])
print('From second element fourth element',arr[1:4])
```

```python
#2D array
arr=np.array([[1,2,3],[4,5,6]])
print('first row',arr[0])
print('first element of first row',arr[0,0])
print('first element of second row',arr[1,0])
print('first two rows',arr[:2])
print('first two rows and first two columns',arr[:2,:2])
```

```python
#boolean indexing
arr=np.array([1,2,3,4,5,6])
mask=arr>3
print('mask',mask)
print('elements greater than 3',arr[mask])
```

```python
#reshaping and combining array
arr=np.array([[1,2,3],[4,5,6]])
print('orginal array',arr)
print('reshaped array',arr.reshape(3,2))
```

```python
#concatenation
a=np.array([[1,2,3],[4,5,6]])
b=np.array([[7,8,9],[10,11,12]])
print('concatenation',np.concatenate((a,b)))
```

```python
#stacking
a=np.array([1,2])
b=np.array([3,4])
print('vertical stacking',np.vstack((a,b)))
print('horizontal stacking',np.hstack((a,b)))
```

```python
#broadcasing
a=np.array([1,2,3])
a+5
```

```python
#saving and loading a file
a=np.array([[1,2,3],[4,5,6]])
np.save('array.npy',a)
b=np.load('array.npy')
print(b)
```
