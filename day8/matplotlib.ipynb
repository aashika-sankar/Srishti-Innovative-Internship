```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y)
plt.show()
```

```python
#plot coustamization
plt.plot(x,y,color='green',linestyle='--',marker='o')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#multiple lines
x=[1,2,3,4]
y1=[1,4,9,16]
y2=[1,8,27,64]
plt.plot(x,y1,label='square')
plt.plot(x,y2,label='cube')
plt.legend()
plt.show()
```

```python
#bar chart
x=[1,2,3,4]
y=[10,20,25,30]
plt.bar(x,y)
plt.show()
```

```python
#histogram
data=[1,1,1,2,2,3,4,5,5,5]
plt.hist(data,bins=5,edgecolor='black')
plt.show()
```

```python
#scatter plot
x=[1,2,3,4]
y=[10,20,25,30]
plt.scatter(x,y)
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot([1,2,3],[1,2,3])
plt.subplot(1,2,2)
plt.plot([1,2,3],[3,2,1])
plt.show()
```

```python
#saving figures
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y)
plt.savefig('plot.png')
```
