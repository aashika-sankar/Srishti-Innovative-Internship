```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Training the Decision Tree Classification model on the
Training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
clf2 = DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
clf2.fit(X_train, y_train)
```

Predicting a new result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
print(clf2.predict(sc.transform([[51,87000]])))
```

Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
y_pred2 = clf2.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm2 = confusion_matrix(y_test, y_pred2)
print(cm2)
accuracy_score(y_test, y_pred2)
```

Visualising the Training set results

```python
from sklearn.inspection import DecisionBoundaryDisplay
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
        classifier,
        X_train,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )

plt.scatter(X_train[:, 0], X_train[:, 1], 
            c=y_train, 
            s=20, edgecolors="k")
plt.title('Decision Tree-Classifier(Training set)')
plt.show()
```

Visualising the Test set results

```python
from sklearn.inspection import DecisionBoundaryDisplay
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
        classifier,
        X_test,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )

plt.scatter(X_test[:, 0], X_test[:, 1], 
            c=y_test, 
            s=20, edgecolors="k")
plt.title('Decision Tree-Classifier(Testing set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
def plot_decision_tree(clf_object, feature_names, class_names):
    plt.figure(figsize=(15, 10))
    tree.plot_tree(clf_object, 
                   filled=True, 
                   feature_names=feature_names, 
                   class_names=class_names, 
                   rounded=True)
    plt.show()
plot_decision_tree(classifier, ['Age','EstimatedSalary'], ['Not Purchased', 'Purchased'])
```

```python
importances = classifier.feature_importances_
print("Feature importances:")
for name, importance in zip(['Age', 'EstimatedSalary'], importances):
    print(f"{name}: {importance:.4f}")
```

```python
plot_decision_tree(clf2, ['Age','EstimatedSalary'], ['Not Purchased', 'Purchased'])
```
