```python
#basic operations
x=10
y=3
print('addition:', x+y)
print('subtraction:', x-y)
print('multiplication:', x*y)
print('division:', x/y)
print('modulus:', x%y)
print('exponentiation:', x**y)
print('floor division:', x//y)
```

```python
#relational operators
print('x > y:', x > y)
print('x >= y:', x >= y)
print('x < y:', x < y)
print('x <= y:', x <= y)
print('x == y:', x == y)
print('x != y:', x != y)
```

```python
#logical operators
print('x > 5 and y < 5:', x > 5 and y < 5)
print('x > 5 or y < 5:', x > 5 or y < 5)
print('not (x > 5):', not (x > 5))
```

```python
#swapping values
a = 5
b = 10
a, b = b, a
print('After swapping: a =', a, ', b =', b)
```

```python
string="hello world"
print('original:',string)
print('length:', len(string))
print('first character:', string[0])
print('sixth character:', string[6])
print('last character:', string[-1])
print('third character from the end:', string[-3])
print('substring (0:5):', string[0:5])
print('substring (6:11):', string[6:11])
print('substring (:5):', string[:5])
print('substring (6:):', string[6:])
print('every second character:', string[::2])
print('uppercase:', string.upper())   
print('lowercase:', string.lower())
print('after replacement:', string.replace("world","python"))
print('count of "o":', string.count("o"))
print('index of "world":', string.find("world"))
print(f'{string} is a string')
```

```python
#do-while loop
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break
    
```

```python
#oops
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
        
number=int(input("Enter a number: "))
per = []
for i in range(number):
    name=input("Enter the name")
    age=int(input('Enter the age'))
    p=Person(name,age)
    
    per.append(p)
    
for i in per:
    i.greet()
```
